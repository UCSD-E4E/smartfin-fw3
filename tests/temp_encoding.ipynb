```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
real_temp = np.linspace(-10,40,200, dtype=np.float32)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(real_temp, (128.0 * (real_temp)).astype(np.int16), label='Dry')
ax.plot(real_temp, (128.0 * (real_temp - 100)).astype(np.int16), label='Wet')
ax.grid()
fig.legend()

ax.set_xlabel('Real (measured) temperature ($^\circ$C)')
ax.set_ylabel('Combined temperature/water')
```
